```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split


# Se divide el conjunto de datos en entrenamiento y prueba (80% para entrenamiento, 20% para prueba)
X_train, X_test, y_train, y_test = train_test_split(df[['fechaalta','EDAD','SEXO','ENTRESIDENCIA','TIPOURGENCIA','ENVIADOA','AFECPRIN','fechaingreso','MES_ESTADISTICO']], df['MOTATE'], test_size=0.2, random_state=42)

```

```python
# Se selecciona un subconjunto representativo de datos (puedes ajustar la proporciÃ³n segÃºn sea necesario)
X_subset = X_train.sample(frac=0.2, random_state=42)

```

```python
# Escalar las caracterÃ­sticas para mejorar el rendimiento del modelo
scaler = StandardScaler()
X_subset_scaled = scaler.fit_transform(X_subset)

```

```python
# Inicializar el modelo DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Ajusta los parÃ¡metros segÃºn tus necesidades
dbscan_labels = dbscan.fit_predict(X_subset_scaled)

# AÃ±adir la etiqueta de cluster a tu conjunto de datos
X_clustered = X_subset.copy()
X_clustered['Cluster'] = dbscan_labels
```

```python
##NO EJECUTAR

# Visualizar los resultados del clustering
sns.pairplot(X_clustered, hue='Cluster', palette='Dark2', diag_kind='kde')
plt.show()
```

```python
# Contar puntos ruidosos
num_noise_points = sum(1 for label in dbscan.labels_ if label == -1)
print(f"NÃºmero de Puntos Ruidosos: {num_noise_points}")

```

```python
from sklearn.metrics import silhouette_score

# Calcular el Silhouette Score
silhouette_avg = silhouette_score(X_subset_scaled, dbscan.labels_)
print(f"Silhouette Score: {silhouette_avg}")
```
