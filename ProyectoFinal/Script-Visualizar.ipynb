```python
import pandas as pd
import numpy as np
from pandas import DataFrame, read_csv
# Import the standard libraries for the data preprocessing

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = read_csv('URGENCIAS.txt', sep='|', index_col=['ID'])

```

```python
#Cargamos Data Set de Catalogo
dfCat = read_csv('CAT_CIE_10_2021.csv')

#Extraemos claves del Catalogo
dfCat_subset = dfCat[['CLAVE', 'NOMBRE']]

#Lo usamos para Mapear el dataset principal
df['AFECPRIN'] = df['AFECPRIN'].map(dfCat_subset.set_index('CLAVE')['NOMBRE'])

df.head()
```

```python
df.head()

df = df[df.CVEEDAD == 5]
df = df.drop(['CVEEDAD'], axis=1)

```

```python
cross_tab = pd.crosstab(df['EDAD'], df['AFECPRIN'])
cross_tab
```

```python

#Tomamos los valores mas comunes
top_values = df['AFECPRIN'].value_counts().head(50).index
#Filtramos df
filtered_df = df[(df['EDAD'].isin(range(19, 51)))&(df['AFECPRIN'].isin(top_values))]
cross_tab = pd.crosstab(filtered_df['EDAD'], filtered_df['AFECPRIN'])


cross_tab
```

```python
#De ser necesario, exportamos a un csv
cross_tab.to_csv('crosstab_result.csv', index=True)
```
