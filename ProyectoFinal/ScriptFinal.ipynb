```python
dfCat = read_csv('CATSEXO.csv')
dfCat
```

```python
#Extraemos claves del Catalogo de Entidades
dfCat_subset = dfCat[['IdSexo', 'Descrip']]

#Lo usamos para Mapear el dataset principal
df['SEXO'] = df['SEXO'].map(dfCat_subset.set_index('IdSexo')['Descrip'])
df.head()
```

```python
dfCat = read_csv('CatMotAtencionUrg.csv')
dfCat.head()
```

```python
#Extraemos claves del Catalogo de Motivo de Atención
dfCat_subset = dfCat[['IDMotAte', 'Descrip']]

#Lo usamos para Mapear el dataset principal
df['MOTATE'] = df['MOTATE'].map(dfCat_subset.set_index('IDMotAte')['Descrip'])
df.head()
```

```python
dfCat = read_csv('CatTipoUrgencia.csv')
dfCat.head()
```

```python
#Extraemos claves del Catalogo de Tipo de Urgencia
dfCat_subset = dfCat[['IdTipoUrgencia', 'Descrip']]

#Lo usamos para Mapear el dataset principal
df['TIPOURGENCIA'] = df['TIPOURGENCIA'].map(dfCat_subset.set_index('IdTipoUrgencia')['Descrip'])
df.head()
```

```python
dfCat = read_csv('CatEnviado.csv')
dfCat.head()
```
