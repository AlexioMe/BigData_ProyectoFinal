Vemos que los valores únicos en la columna Afeprin son demasiados por lo que se tomará una muestra menor

```python
#Extraemos las 10 edades mas comunes
top_edades = df[df['EDAD'] >= df['EDAD'].quantile(0.9)]

top_edades.info()

```

```python
#Extraemos los 25 valores mas populares
top_afecprin = df['AFECPRIN'].value_counts().nlargest(25).index

# Hacemos un subset con los valores mas populares
df_top = df[df['AFECPRIN'].isin(top_afecprin)]
plt.figure(figsize=(12, 6))
#Creamos el histograma usando unicamente los 10 primeros caracteres
plt.hist(df_top['AFECPRIN'].apply(lambda x: x[:16]), bins=25, edgecolor='black')

#Ponemos etiquetas
plt.xlabel("Afectación")
plt.ylabel("Frecuencia")

#Rotamos etiquetas
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python

plt.figure(figsize=(12, 6))

#Declaramos Grafico de dispersion
plt.scatter(df_top['AFECPRIN'].apply(lambda x: x[:16]), df_top['EDAD'], color='blue', label='EDAD')
plt.xlabel("Aflicción")
plt.ylabel("EDAD")
plt.xticks(rotation=45, ha="right")
plt.legend()

plt.show()
```
